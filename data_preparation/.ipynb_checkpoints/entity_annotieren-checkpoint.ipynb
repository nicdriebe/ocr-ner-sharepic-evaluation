### Script zum Annotieren der Entitäten 

Die "label" wurden manuell in der Json Datei eingetragen  
* Der Code wurde hauptsächlich von ChatGPT-4o generiert und nur leicht angepasst.

```python
import json
import re

def generate_entity_candidates(json_file, output_file):
    with open(json_file, "r", encoding="utf-8") as f:
        data = json.load(f)

    annotated = []

    for entry in data:
        file_name = entry["file_name"]
        lines = entry["lines"]
        entity_list = []

        for i, line in enumerate(lines):
            for match in re.finditer(r"\S+", line):  # alle "Wörter"
                start_char = match.start()
                end_char = match.end()
                token = match.group()

                entity_list.append({
                    "text": token,
                    "start_line": i,
                    "start_char": start_char,
                    "end_char": end_char,
                    "label": ""  # manuell ergänzen
                })

        annotated.append({
            "file_name": file_name,
            "lines": lines,
            "entities": entity_list
        })

    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(annotated, f, indent=2, ensure_ascii=False) # ascii false, damit Umlaute erhalten bleiben

```

```python
input_file = "original_text.json"
output_file = "entity_annotiert.json"

generate_entity_candidates(input_file, output_file)
```
