### Erstellung der CSV Datei / Befüllen der Spalten

* die hier aufgerufenen CSV Dateien "img.csv", "img2.csv", "bilder_neu.csv", "sharepics.csv" wurden im Verlauf der Arbeit umbenannt und im Projektverzeichnis verschoben. DSie sind nicht mehr Teil des Projekts, aber waren die Grundlage für die finale Datei "images_data_with_groundtruth.csv

#### Erstellen der Spalten und befüllen mit Bildname, Bildordnername (Kategorie)

```python
# dieser Codeblock wurde von ChatGPT generiert und nur leicht modifiziert
import os
import csv


images = "./bilder"  

# Ziel-CSV-Datei
csv_datei = "img.csv"

# Bildzähler
zaehler = 1

# CSV vorbereiten
with open(csv_datei, mode='w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["orignal name", "new name", "catgeory"])

    # Durchlaufe alle Unterordner
    for root, dirs, files in os.walk(images):
        for datei in files:
            if datei.lower().endswith((".jpg", ".jpeg", ".png")):
                alter_pfad = os.path.join(root, datei)
                relative_pfad = os.path.relpath(alter_pfad, images)

                # Kategorie = Name des Unterordners (direkt unter "bilder")
                kategorie = os.path.relpath(root, images).split(os.sep)[0]

                # Neuer Dateiname: nur Nummer mit gleicher Endung
                dateiendung = os.path.splitext(datei)[1].lower()
                neuer_dateiname = f"{zaehler:04d}{dateiendung}"
                neuer_pfad = os.path.join(root, neuer_dateiname)
                relative_neu = os.path.relpath(neuer_pfad, images)

                # Umbenennen
                os.rename(alter_pfad, neuer_pfad)


                # In CSV speichern
                csv_writer.writerow([relative_pfad, os.path.relpath(neuer_pfad, images), kategorie])
                zaehler += 1



```

```python
import os

ordnerpfad = "bilder_neu"

for _, row in df.iterrows():
    original = row["original name"]
    neu = row["new name"]
    
    alter_pfad = os.path.join(ordnerpfad, original)
    neuer_pfad = os.path.join(ordnerpfad, neu)
    
    if os.path.exists(alter_pfad):
        os.rename(alter_pfad, neuer_pfad)
        print(f"{original} ➜ {neu}")
    else:
        print(f"Datei nicht gefunden: {original}")
```
