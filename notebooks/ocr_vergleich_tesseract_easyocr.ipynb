# Vergleich der Ergebnisse Tesseract und EasyOCR

Dieses notebook bietet eine Ãœbersicht und visuelle Darstellung der Ergebnisse der Evaluation von Tesseract und EasyOCR im Vergleich

---

##  CER / WER / Erkennungsrate - zeilenbasiert 

**Tesseract (psm11):**    
Erkennungsrate: 0.498  
CER: 0.546  
WER: 0.675  


**EasyOCR:**    
Erkennungsrate: 0.722  
CER: 0.338  
WER: 0.525  


```python
import matplotlib.pyplot as plt
import numpy as np

# Daten
metrics = ['Erkennungsrate', 'CER', 'WER']
tesser = [0.498, 0.546, 0.675]  # mit nicht erkannten Zeilen
easy = [0.722, 0.338, 0.525]  # nur gematchte Zeilen

x = np.arange(len(metrics))
width = 0.35

# Plot
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, tesser, width, label='Tesseract', color='skyblue')
bars2 = ax.bar(x + width/2, easy, width, label='EasyOCR', color='#527a68')


# Beschriftung
ax.set_ylabel('Wert')
ax.set_title('Tesseract (psm11) vs. EasyOCR - alle Zeilen')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylim(0, 1)

# Werte anzeigen
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.3f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
