in diesem notebook werden die Werte der Evaluation der einzelne "PSM-Werte" verglichen.


---
#### Vergleich Zeilen-Erkennungsrate und durchschnittlicher CER / WER
psm3, psm1, psm11

```python
import matplotlib.pyplot as plt
import numpy as np

# Daten
metrics = ['Erkennungsrate', 'CER', 'WER']
psm11 = [0.498, 0.546, 0.675]  
psm1 = [0.418, 0.618, 0.718] 
psm = [0.420, 0.617, 0.717]

# Abstand zwischen den Metriken vergrößern
x = np.array([0, 1.5, 3])
width = 0.25

x = np.arange(len(metrics))


# Plot
fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(x - width,psm,  width, label='psm 3',  color='#69A9E1')
bars2 = ax.bar(x,psm1, width, label='psm 1',  color='#F1872B')
bars3 = ax.bar(x + width,psm11, width, label='psm 11', color='#FFC600')



# Beschriftung
ax.set_ylabel('Wert')
ax.set_title('Vergleich psm 1 psm 3 psm 11 - CER / WER / Erkennunsgrate ')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylim(0, 1)

# Werte anzeigen
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

#### Vergleich CER pro Bildkategorie
psm3, psm1, psm11

```python
import matplotlib.pyplot as plt
import pandas as pd

# Kategorien und Werte
categories = ['simple', 'noisy', 'skewed', 'fancy', 'photo']
cer_psm11 = [36.19, 70.55, 54.33, 65.15, 62.40] 
cer_psm1 = [48.24, 71.17, 63.95, 68.07, 71.66]          
cer = [48.24, 70.08, 64.17, 68.08, 71.74]          



# DataFrame
df = pd.DataFrame({
    'category': categories,
    'CER_p3': cer,
    'CER_p1': cer_psm1,
    'CER_p11': cer_psm11,
})

# Plot
fig, ax = plt.subplots(figsize=(10, 6))

x = range(len(categories))
bar_width = 0.30

# Balken zeichnen
bars1 = ax.bar([i - bar_width for i in x], df['CER_p3'], width=bar_width, label='CER mit psm 3 (%)', color='#69A9E1')
bars2 = ax.bar(x, df['CER_p1'], width=bar_width, label='CER mit psm 1 (%)', color='#F1872B')
bars3 = ax.bar([i + bar_width for i in x], df['CER_p11'], width=bar_width, label='CER mit psm 11 (%)', color='#FFC600')

# Achsen und Beschriftungen
ax.set_ylabel('CER (%)')
ax.set_ylim(0, 100)
ax.set_xlabel('Bildkategorie')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend(loc='upper right')

plt.title('CER pro Bildkategorie / Vergleich psm3 psm1 psm11 / zeilenbasiert')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Werte anzeigen
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### Vergleich psm3 und psm 11 - Qualität der erkannten der Zeilen - CER / WER
CER und WER wird nur auf mit dem Ground Truth gematchten Zeilen berrechnet   
(Erkennungsrate verändert sich dadurch jeweils nicht, wird aber dennoch angegeben)

```python
import matplotlib.pyplot as plt
import numpy as np

# Daten
metrics = ['Erkennungsrate', 'CER/only matched lines', 'WER/only matched lines']
psm11 = [0.498, 0.105, 0.395]  
psm = [0.420, 0.107, 0.389]



x = np.arange(len(metrics))
width = 0.35

# Plot
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, psm, width, label='psm 3', color='#69A9E1')
bars2 = ax.bar(x + width/2, psm11, width, label='psm 11', color='#FFC600')


# Beschriftung
ax.set_ylabel('Wert')
ax.set_title('Vergleich psm 3 psm 11 - CER / WER - nur gematchte Zeilen')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylim(0, 1)

# Werte anzeigen
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.3f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
