# EasyOCR - Evaluation - 02 / 0 - Preprocessing

In diesem Notebook wird die Auswirkung vom Preprocessing der Bilder auf die Ergebnisse der Textextraktion durch EasyOCR untersucht.
Dazu werden die bilder verschiedenen Stufen der Bildbearbeitung unterzogen. Dies geschieht mit OpenCV, ein Tool mit dem Kontrast Helligkeit  verändert werden kann oder Noisereduktion durch Blureffekte erzielt werden kann.   
Um Verbesserung zu erzielen wird der "Confidence" Wert der von EasyOCR bei jeder Textextraktion ausgegeben als Maßstab für die verbesserte Textextraktion herangezogen. Bei einem Confidence Wert von mindestens 50% auf allen Zeilen, wird kein weiteres Preprocessing angewendet.

Das Verfahren des Preprocessings wird zunächst auf einem einzelnen Bild getestet.

```python
# Funktion zur Evaluation - "ocr_eval"
import os
import json
from PIL import Image
import pandas as pd
import numpy as np


import sys # ermöglicht Zugriff auf übergeordnetes Verzeichnis
sys.path.append(os.path.abspath("../../utils"))
from funktionen import cer_for_matching, match_gt_to_ocr, calculate_char_accuracy, calculate_word_accuracy, calculate_wer,calculate_cer, CER_THRESHOLD


### Konstanzen ###
IMG_DIR = "../../data/images/insta_images"
JSON_PATH = "../../data/data_annotated.json"


### Hilfsfunktionen ###

# Textextraktion erfolgt zeilenweise
def extract_ocr_lines(image_path):
    return reader.readtext(image_path, detail=0)


### Hauptprozess ###


def ocr_eval (file_name):

    with open(JSON_PATH, "r", encoding="utf-8") as f:
        all_data = json.load(f)
    
    ergebnisse = []
    
    eintrag = next((item for item in all_data if item["file_name"] == file_name), None)

    
    gt_lines = eintrag.get("lines", [])
    img_path = os.path.join(IMG_DIR, file_name)
    #if not os.path.exists(img_path):
    #    continue
    
    ocr_lines = extract_ocr_lines(img_path)
    matches = match_gt_to_ocr(gt_lines, ocr_lines, CER_THRESHOLD)
    
    total_gt = len(gt_lines)
    total_ocr = len(ocr_lines)
    total_matches = 0
    cer_list, wer_list, char_acc_list, word_acc_list = [], [], [], []
    
    matched_cer = []
    matched_wer = []
    matched_char_acc = []
    matched_word_acc = []
    
    for gt_idx, ocr_idx in matches:
        gt_line = gt_lines[gt_idx]
        if ocr_idx is not None:
            ocr_line = ocr_lines[ocr_idx]
            cer_score = calculate_cer(gt_line, ocr_line)
            wer_score = calculate_wer(gt_line, ocr_line)
            char_acc = calculate_char_accuracy(gt_line, ocr_line)
            word_acc = calculate_word_accuracy(gt_line, ocr_line)
    
            matched_cer.append(cer_score)
            matched_wer.append(wer_score)
            matched_char_acc.append(char_acc)
            matched_word_acc.append(word_acc)
    
            total_matches += 1
    
    ergebnisse.append({
        "file_name": file_name,
        "gt_lines": total_gt,
        "ocr_lines": total_ocr,
        "matched_lines": total_matches,
        "mean_cer": sum(matched_cer) / total_matches if total_matches else None,
        "mean_wer": sum(matched_wer) / total_matches if total_matches else None,
        "mean_char_acc": sum(matched_char_acc) / total_matches if total_matches else None,
        "mean_word_acc": sum(matched_word_acc) / total_matches if total_matches else None
    })
    
    
    df_results = pd.DataFrame(ergebnisse)
    display(df_results)
```

```python
# Preprocessing Funktion - "optimize_ocr"
import cv2
import numpy as np
import easyocr
import json

reader = easyocr.Reader(['de'])

def preprocess_variants(gray_img, alpha=1.0, beta=0, invert=False, binarize=False):
    proc = cv2.convertScaleAbs(gray_img, alpha=alpha, beta=beta)
    if invert:
        proc = cv2.bitwise_not(proc)
    if binarize:
        _, proc = cv2.threshold(proc, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return proc

def run_ocr(img):
    return reader.readtext(img)

def bbox_iou(boxA, boxB):
    """Berechnet IOU (Intersection over Union) zweier Bounding Boxes."""
    # Vereinfacht: nutze umschreibendes Rechteck für beide Polygone
    xA1, yA1 = np.min(boxA, axis=0)
    xA2, yA2 = np.max(boxA, axis=0)
    xB1, yB1 = np.min(boxB, axis=0)
    xB2, yB2 = np.max(boxB, axis=0)

    xI1, yI1 = max(xA1, xB1), max(yA1, yB1)
    xI2, yI2 = min(xA2, xB2), min(yA2, yB2)

    interArea = max(0, xI2 - xI1) * max(0, yI2 - yI1)
    boxAArea = (xA2 - xA1) * (yA2 - yA1)
    boxBArea = (xB2 - xB1) * (yB2 - yB1)
    unionArea = boxAArea + boxBArea - interArea

    if unionArea == 0:
        return 0
    return interArea / unionArea

def merge_best_lines(existing_results, new_results, iou_threshold=0.5):
    """Speichert pro Zeile (Bounding Box) nur die Version mit der höchsten Confidence."""
    for new in new_results:
        new_box = new['bbox']
        new_conf = new['confidence']

        replaced = False
        for i, existing in enumerate(existing_results):
            iou = bbox_iou(np.array(existing['bbox']), np.array(new_box))
            if iou > iou_threshold:
                if new_conf > existing['confidence']:
                    existing_results[i] = new
                replaced = True
                break

        if not replaced:
            existing_results.append(new)

def save_results(results, filepath):
    with open(filepath, 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

def optimize_ocr(img_path, save_path, max_attempts=50):
    img = cv2.imread(img_path)
    if img is None:
        raise FileNotFoundError(f"Bild nicht gefunden: {img_path}")

    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    best_results = []

    attempts = 0
    #for alpha in [0.8, 1.0, 1.2, 1.5]: # Parameter des 1.Durchlauf
     #   for beta in [-50, 0, 50]:
    for alpha in [1.0, 3.0, 5.0, 8.0]: # Parameter des 2.Durchlauf
        for beta in [-20, 0, 10, 25]:
            for invert in [False, True]:
                for binarize in [False, True]:
                    if attempts >= max_attempts:
                        save_results(best_results, save_path)
                        return best_results

                    proc_img = preprocess_variants(
                        gray_img, alpha=alpha, beta=beta,
                        invert=invert, binarize=binarize
                    )
                    ocr_output = run_ocr(proc_img)
                    results = []
                    for r in ocr_output:
                        box = [[float(coord[0]), float(coord[1])] for coord in r[0]]  # Box: float64 → float # NumPy-Float (float64 oder float32) 
                        text = r[1]                                                                          # in einen normalen Python-Float umgewandelt
                        conf = float(r[2])  # Confidence: np.float32 → float
                    
                        results.append({
                            "text": text,
                            "bbox": box,
                            "confidence": conf,
                        })

                    merge_best_lines(best_results, results)
                    attempts += 1

    save_results(best_results, save_path)
    return best_results
```

```python
# sortieren des preprocessing outputs in lines - einfach
def group_texts_to_lines1(ocr_output, y_threshold=20):
    # Konvertiert bbox-Textobjekte in Zeilen durch Gruppierung entlang der Y-Achse
    lines = []
    used = set()
    
    # Extrahiere relevante Infos
    entries = []
    for entry in ocr_output:
        if not entry['text'].strip():
            continue
        y_center = np.mean([point[1] for point in entry['bbox']])
        x_start = min(point[0] for point in entry['bbox'])
        entries.append({
            'text': entry['text'],
            'y': y_center,
            'x': x_start
        })
    
    # Sortiere nach Y, dann nach X
    entries.sort(key=lambda e: (e['y'], e['x']))

    while entries:
        base = entries.pop(0)
        current_line = [base]
        remaining = []
        for e in entries:
            if abs(e['y'] - base['y']) < y_threshold:
                current_line.append(e)
            else:
                remaining.append(e)
        # Sortiere innerhalb der Zeile nach X
        current_line.sort(key=lambda e: e['x'])
        line_text = " ".join([e['text'] for e in current_line])
        lines.append(line_text)
        entries = remaining

    return lines
```

```python
# sortieren des preprocessing outputs in lines - keine doppelungen in boxes "group_texts_to_lines2"
import numpy as np

def iou(box1, box2):
    # Berechne Intersection over Union zweier Bboxen
    x0_1, y0_1 = np.min(box1, axis=0)
    x1_1, y1_1 = np.max(box1, axis=0)

    x0_2, y0_2 = np.min(box2, axis=0)
    x1_2, y1_2 = np.max(box2, axis=0)

    xi0 = max(x0_1, x0_2)
    yi0 = max(y0_1, y0_2)
    xi1 = min(x1_1, x1_2)
    yi1 = min(y1_1, y1_2)

    inter_area = max(0, xi1 - xi0) * max(0, yi1 - yi0)
    box1_area = (x1_1 - x0_1) * (y1_1 - y0_1)
    box2_area = (x1_2 - x0_2) * (y1_2 - y0_2)

    union_area = box1_area + box2_area - inter_area
    if union_area == 0:
        return 0
    return inter_area / union_area

def group_texts_to_lines2(ocr_output, y_threshold=20, confidence_threshold=0.98, iou_threshold=0.01):
    # 1. Confidence-Filter
    filtered = [
        e for e in ocr_output
        if e['confidence'] >= confidence_threshold and e['text'].strip()
    ]

    # 2. Extrahiere Text + Koordinaten
    entries = []
    for entry in filtered:
        y_center = np.mean([pt[1] for pt in entry['bbox']])
        x_start = min(pt[0] for pt in entry['bbox'])
        entries.append({
            'text': entry['text'],
            'y': y_center,
            'x': x_start,
            'bbox': np.array(entry['bbox']),
        })

    # 3. Sortiere nach y, dann x
    entries.sort(key=lambda e: (e['y'], e['x']))

    lines = []
    while entries:
        base = entries.pop(0)
        current_line = [base]
        remaining = []

        for e in entries:
            same_line = abs(e['y'] - base['y']) < y_threshold
            overlap = any(iou(e['bbox'], other['bbox']) > iou_threshold for other in current_line)
            if same_line and not overlap:
                current_line.append(e)
            else:
                remaining.append(e)

        current_line.sort(key=lambda e: e['x'])
        lines.append(" ".join(e['text'] for e in current_line))
        entries = remaining

    return lines
```

```python
import os
import json
from PIL import Image
import pandas as pd

import sys # ermöglicht Zugriff auf übergeordnetes Verzeichnis
sys.path.append(os.path.abspath("../../utils"))
from funktionen import cer_for_matching, match_gt_to_ocr, calculate_char_accuracy, calculate_word_accuracy, calculate_wer,calculate_cer, CER_THRESHOLD




# === Konfiguration ===
IMG_DIR = "../../data/images/insta_images"
JSON_PATH = "../../data/data_annotated.json"
JSON_BASE_PATH_OCR = "../../data/OCR/easyocr/preprocessed/results_preprocessing"



### HILFSFUNKTIONEN ###

def extract_ocr_lines(image_path):
    return reader.readtext(image_path, detail=0)


# === Hauptprozess ===

with open(JSON_PATH, "r", encoding="utf-8") as f:
    all_data = json.load(f)

ergebnisse = []

for eintrag in all_data:
    file_name = eintrag.get("file_name")
    if not file_name:
        continue   

    eintrag = next((item for item in all_data if item["file_name"] == file_name), None)

    base_name, _ = os.path.splitext(file_name)
    json_ocr = os.path.join(JSON_BASE_PATH_OCR, base_name + ".json")
    with open(json_ocr, "r", encoding="utf-8") as f:
        ocr_data = json.load(f)

    gt_lines = eintrag.get("lines", [])
    img_path = os.path.join(IMG_DIR, file_name)
    if not os.path.exists(img_path):
        continue

    ocr_lines = group_texts_to_lines2(ocr_data, y_threshold=20)
    matches = match_gt_to_ocr(gt_lines, ocr_lines, CER_THRESHOLD)

    total_gt = len(gt_lines)
    total_ocr = len(ocr_lines)
    total_matches = 0
    cer_list, wer_list, char_acc_list, word_acc_list = [], [], [], []

    for gt_idx, ocr_idx in matches:
        gt_line = gt_lines[gt_idx]
        if ocr_idx is not None:
            ocr_line = ocr_lines[ocr_idx]
            cer_score = calculate_cer(gt_line, ocr_line)
            wer_score = calculate_wer(gt_line, ocr_line)
            char_acc = calculate_char_accuracy(gt_line, ocr_line)
            word_acc = calculate_word_accuracy(gt_line, ocr_line)

            cer_list.append(cer_score)
            wer_list.append(wer_score)
            char_acc_list.append(char_acc)
            word_acc_list.append(word_acc)
            total_matches += 1
        else:
            cer_list.append(1.0)
            wer_list.append(1.0)
            char_acc_list.append(0.0)
            word_acc_list.append(0.0)

    ergebnisse.append({
        "file_name": file_name,
        "gt_lines": total_gt,
        "ocr_lines": total_ocr,
        "matched_lines": total_matches,
        "mean_cer": sum(cer_list) / total_gt if total_gt else None,
        "mean_wer": sum(wer_list) / total_gt if total_gt else None,
        "mean_char_acc": sum(char_acc_list) / total_gt if total_gt else None,
        "mean_word_acc": sum(word_acc_list) / total_gt if total_gt else None,
        "ocr_easyocr": ocr_lines
    })

df_results_line = pd.DataFrame(ergebnisse)
display(df_results_line)
```

```python
import os
import json
from PIL import Image
import pandas as pd
import numpy as np


import sys # ermöglicht Zugriff auf übergeordnetes Verzeichnis
sys.path.append(os.path.abspath("../../utils"))
from funktionen import cer_for_matching, match_gt_to_ocr, calculate_char_accuracy, calculate_word_accuracy, calculate_wer,calculate_cer, CER_THRESHOLD


### Konstanzen ###
IMG_DIR = "../../data/images/insta_images"
JSON_PATH = "../../data/data_annotated.json"
JSON_BASE_PATH_OCR = "../../data/OCR/easyocr/preprocessed/results_preprocessing"


### Hauptprozess ###


def ocr_eval_preproc():

    with open(JSON_PATH, "r", encoding="utf-8") as f:
        all_data = json.load(f)

    ergebnisse = []

    for eintrag in all_data:
        file_name = eintrag.get("file_name")
        if not file_name:
            continue   
    
        eintrag = next((item for item in all_data if item["file_name"] == file_name), None)
    
        base_name, _ = os.path.splitext(file_name)
        json_ocr = os.path.join(JSON_BASE_PATH_OCR, base_name + ".json")
        with open(json_ocr, "r", encoding="utf-8") as f:
            ocr_data = json.load(f)
        
        gt_lines = eintrag.get("lines", [])
        img_path = os.path.join(IMG_DIR, file_name)
        if not os.path.exists(img_path):
            continue
        
        ocr_lines = group_texts_to_lines2(ocr_data, y_threshold=20)
        
        matches = match_gt_to_ocr(gt_lines, ocr_lines, CER_THRESHOLD)
        
        
        total_gt = len(gt_lines)
        total_ocr = len(ocr_lines)
        total_matches = 0
        cer_list, wer_list, char_acc_list, word_acc_list = [], [], [], []
        
        matched_cer = []
        matched_wer = []
        matched_char_acc = []
        matched_word_acc = []
        
        for gt_idx, ocr_idx in matches:
            gt_line = gt_lines[gt_idx]
            if ocr_idx is not None:
                ocr_line = ocr_lines[ocr_idx]
                cer_score = calculate_cer(gt_line, ocr_line)
                wer_score = calculate_wer(gt_line, ocr_line)
                char_acc = calculate_char_accuracy(gt_line, ocr_line)
                word_acc = calculate_word_accuracy(gt_line, ocr_line)
        
                matched_cer.append(cer_score)
                matched_wer.append(wer_score)
                matched_char_acc.append(char_acc)
                matched_word_acc.append(word_acc)
        
                total_matches += 1
                
       
        ergebnisse.append({
            "file_name": file_name,
            "gt_lines": total_gt,
            "ocr_lines": total_ocr,
            "matched_lines": total_matches,
            "mean_cer": sum(matched_cer) / total_matches if total_matches else None,
            "mean_wer": sum(matched_wer) / total_matches if total_matches else None,
            "mean_char_acc": sum(matched_char_acc) / total_matches if total_matches else None,
            "mean_word_acc": sum(matched_word_acc) / total_matches if total_matches else None
        })
    
    
    df_results = pd.DataFrame(ergebnisse)
    display(df_results)

    output_path = "../../data/OCR/easyocr/easyocr_evaluation_02_0_preprocessing_matchted_lines_only.csv"
    df_results.to_csv(output_path, index=False, encoding="utf-8")
    print(f"Ergebnisse gespeichert unter: {output_path}")
```

### Entitäten erkannt?

```python
import json
import pandas as pd

JSON_PATH = "../../data/data_annotated.json"
CSV_PATH = "../../data/OCR/easyocr/easyocr_evaluation_02_0_preprocessing_linewise.csv"
OUTPUT_PATH = "../../data/OCR/easyocr/easyocr_evaluation_preprocessing_entity.csv"

# JSON & CSV laden
with open(JSON_PATH, "r", encoding="utf-8") as f:
    gt_data = json.load(f)

ocr_df = pd.read_csv(CSV_PATH)

# NaN in 'ocr_text' durch leeren String ersetzen
ocr_df["ocr_easyocr"] = ocr_df["ocr_easyocr"].fillna("")

# Alle Labels, die wir tracken wollen
labels_to_track = ["EVENT", "TOPIC", "DATE", "TIME", "LOC"]

# Ergebnisliste
results = []

for gt_entry in gt_data:
    file_name = gt_entry["file_name"]
    gt_entities = gt_entry.get("entities", [])

    # OCR-Text zur Datei holen
    ocr_row = ocr_df[ocr_df["file_name"] == file_name]
    if ocr_row.empty:
        print(f"OCR-Text für {file_name} nicht gefunden – übersprungen.")
        continue
    ocr_text = ocr_row.iloc[0]["ocr_easyocr"]

    # Initiale Zähler
    result_row = {"file_name": file_name}
    for label in labels_to_track:
        result_row[f"Anzahl_{label}_gt"] = 0
        result_row[f"Anzahl_{label}_erkannt"] = 0

    erkannte_entitaeten = []
    nicht_erkannte_entitaeten = []

    for ent in gt_entities:
        label = ent["label"]
        text = ent["text"]

        if label not in labels_to_track:
            continue

        # Ground-Truth-Zähler erhöhen
        result_row[f"Anzahl_{label}_gt"] += 1

        # Prüfen, ob Text im OCR vorkommt
        if text in ocr_text:
            result_row[f"Anzahl_{label}_erkannt"] += 1
            erkannte_entitaeten.append(f"[{label}] {text}")
        else:
            nicht_erkannte_entitaeten.append(f"[{label}] {text}")

    # Listen speichern als Strings
    result_row["erkannte_entities"] = "; ".join(erkannte_entitaeten)
    result_row["nicht_erkannte_entities"] = "; ".join(nicht_erkannte_entitaeten)

    results.append(result_row)

# In DataFrame umwandeln und speichern
df_result = pd.DataFrame(results)
df_result.to_csv(OUTPUT_PATH, index=False, encoding="utf-8")

print(f"Auswertung gespeichert in: {OUTPUT_PATH}")
```

```python
import pandas as pd


def berechne_erkennungsraten(csv_datei):
    # CSV laden
    df = pd.read_csv(csv_datei)

    # Entitäten, für die Erkennungsraten berechnet werden sollen
    entities = ["EVENT", "TOPIC", "DATE", "TIME", "LOC"]

    # Für jede Entität die Erkennungsrate berechnen
    for ent in entities:
        gt_col = f"Anzahl_{ent}_gt"
        erk_col = f"Anzahl_{ent}_erkannt"
        rate_col = f"Erkennungsrate_{ent}"

        # Division mit Schutz vor 0
        df[rate_col] = df[erk_col] / df[gt_col]
        df[rate_col] = df[rate_col].replace([float("inf"), float("-inf")], 0).fillna(0)

    return df
```

```python
df_ent = berechne_erkennungsraten("../../data/OCR/easyocr/easyocr_evaluation_preprocessing_entity.csv")

df_ent.head(1)
```

```python
mean_reg_rate_Event = df_ent["Erkennungsrate_EVENT"].mean()
mean_reg_rate_Topic = df_ent["Erkennungsrate_TOPIC"].mean()
mean_reg_rate_Date = df_ent["Erkennungsrate_DATE"].mean()
mean_reg_rate_Time = df_ent["Erkennungsrate_TIME"].mean()
mean_reg_rate_Loc = df_ent["Erkennungsrate_LOC"].mean()

reg_rate_ALL = (mean_reg_rate_Event + mean_reg_rate_Topic + mean_reg_rate_Date + mean_reg_rate_Time + mean_reg_rate_Loc) / 5


print(f"Erkennungsrate gesamt: {reg_rate_ALL:.3f}")
print(f"Erkennungsrate pro Entität:")
print(f"Event: {mean_reg_rate_Event:.3f}")
print(f"Topic: {mean_reg_rate_Topic:.3f}")
print(f"Date: {mean_reg_rate_Date:.3f}")
print(f"Time: {mean_reg_rate_Time:.3f}")
print(f"Loc: {mean_reg_rate_Loc:.3f}")
```

### Vergleich - CER / WER / Erkennungsrate mit und ohne Preprocessing

```python
import matplotlib.pyplot as plt
import numpy as np

# Daten
metrics = ['Erkennungsrate', 'CER', 'WER']
easy = [0.722, 0.338, 0.525] 
easy_preproc = [0.465, 0.566, 0.631]  


x = np.arange(len(metrics))
width = 0.35

# Plot
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, easy, width, label='OHNE Preprocessing', color='pink')
bars2 = ax.bar(x + width/2, easy_preproc, width, label='MIT Preprocessing', color='blue')


# Beschriftung
ax.set_ylabel('Wert')
ax.set_title('EasyOCR mit und ohne Preprocessing - zeilenweiser Vergleich mit ground truth')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylim(0, 1)

# Werte anzeigen
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
