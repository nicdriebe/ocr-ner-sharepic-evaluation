- True Positives

### False Positives - Beispielausgabe nach Entität

#### 1. Entität EVENT

```python
import pandas as pd
import json


with open("../../data/NER/gemma2/results_gemma.json", encoding="utf-8") as f:
    pred_data = json.load(f)


with open("../../data/data_annotated.json", encoding="utf-8") as f:
    gold_data = json.load(f)

# Indexiere Golddaten nach file_name
gold_index = {entry["file_name"]: entry for entry in gold_data}
relevant_label = "EVENT"

false_positives_all = []

for pred in pred_data:
    file_name = pred["file_name"]
    fps = [e for e in pred.get("false_positives", []) if e["label"] == relevant_label]

    gold_entry = gold_index.get(file_name, {})
    full_text = gold_entry.get("text", "") 

    
    gold_event_texts = [
        e["text"]
        for e in gold_entry.get("entities", [])
        if e["label"] == relevant_label
    ]

    for fp in fps:
        fragment = fp["text"]

        fp_entry = {
            "file_name": file_name,
            "text_fragment": fragment,
            "start": None,
            "end": None,
            "gold_event_texts": gold_event_texts,
            "is_false_positive": fragment not in gold_event_texts
        }

        false_positives_all.append(fp_entry)


df_fp_event = pd.DataFrame(false_positives_all)
df_fp_event = df_fp_event[["file_name", "text_fragment", "gold_event_texts"]]

pd.set_option('display.max_colwidth', None)


df_fp_event.head(30)
```

```python
df_fp_event.sample(30)[["file_name", "text_fragment", "gold_event_texts"]]
```
