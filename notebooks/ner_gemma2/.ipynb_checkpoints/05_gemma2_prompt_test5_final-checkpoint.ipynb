### Auf Datensatz iterieren und Output im Json-Format speichern

````python
# Funktion extract_entities
import json
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline


# Model und Pipeline
model_name = "google/gemma-2b-it"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device=-1,
    do_sample=False # weniger kreativ, weniger Zufall
)

def extract_entities(data):
    results = []
    
    for eintrag in data:
        file_name = eintrag["file_name"]
        text = eintrag["text"]
    
        full_prompt = """<start_of_turn>user
    Du bist ein Modell zur strukturierten Informationsextraktion aus deutschsprachigen Veranstaltungstexten. 
    
    Extrahiere, wenn möglich, die folgenden Informationen:
    
    - **EVENT** (Art der Veranstaltung)
    - **TOPIC** (Thema der Veranstaltung)
    - **DATE** (Datumsangaben)
    - **TIME** (Uhrzeitangaben)
    - **LOC** (Ort der Veranstaltung, z. B. Straßen mit Hausnummern, Plätze, Gebäude, Stadtteile, Städte)
    
    Regeln:
    - Gib nur Informationen an, bei denen du dir sicher bist.
    - Verwende **ausschließlich Wörter aus dem Originaltext**.
    - Nicht alle Felder müssen vorhanden sein.
    - Mehrfache Einträge pro Feld sind möglich.
    
    Gib das Ergebnis exakt im folgenden Format zurück:
    
    {{
      "EVENT": [...],
      "TOPIC": [...],
      "DATE": [...],
      "TIME": [...],
      "LOC": [...]
    }}
    
    
    
    Hier ist der Text:
    {text}
    
    <end_of_turn>
    <start_of_turn>model
    """
        
        try:
            prompt = full_prompt.format(text=text)
            
            
            output = generator(prompt, max_new_tokens=400)[0]["generated_text"]
            extracted_json = output.split("model")[-1].split("model")[0].strip()  # nur Output nach "model" wird genommen
            extracted = json.loads(extracted_json)
            print("==== extracted ====")
            print(extracted)
        
            results.append({
                "file_name": file_name,
                "entities": extracted
            })
        
        except Exception as e:
            print(f"Fehler bei {file_name}: {e}")
            results.append({
                "file_name": file_name,
                "entities": None,
                "error": str(e)
            })
        
    return results


````

````python
# Datei bereingigen / Nulleinträge entfernen
import json


# JSON laden
with open("../../data/NER/gemma2/prompt5/gemma2_entity_output_prompt5_2.json", "r", encoding="utf-8") as f:
    data = json.load(f)

bereinigt = []
for eintrag in data:
    entities = eintrag.get("entities")

    if not entities or not isinstance(entities, dict):
        continue  # überspringen, wenn kaputt oder leer
    
    # null-Werte rausfiltern 
    entities_cleaned = {
        k: v
        for k, v in entities.items()
        if v is not None
    }

    # überspringen, wenn nach dem Filtern nichts übrig ist (optional)
    if not entities_cleaned:
        continue

    eintrag["entities"] = entities_cleaned
    bereinigt.append(eintrag)

# Speichern
with open("../../data/NER/gemma2/prompt5/gemma2_entity_output_prompt5_2_cleaned.json", "w", encoding="utf-8") as f:
    json.dump(bereinigt, f, ensure_ascii=False, indent=2)
````

````python
# Json in Struktur des Goldstandard umwandeln

import json

# Datei laden
with open("../../data/NER/gemma2/prompt5/gemma2_entity_output_prompt5_2_cleaned.json", encoding="utf-8") as f:
    gemma_data = json.load(f)

# Normalisieren
def normalize_to_list(value):
    if isinstance(value, list):
        result = []
        for v in value:
            if isinstance(v, str) and v.strip():
                result.extend([item.strip() for item in v.split(",") if item.strip()])
        return result
    elif isinstance(value, str) and value.strip():
        return [item.strip() for item in value.split(",") if item.strip()]
    else:
        return []

# In Goldstandard-ähnliche Struktur bringen
converted = []

for entry in gemma_data:
    file_name = entry.get("file_name")
    raw_entities = entry.get("entities", {})
    
    entities = []
    for label, value in raw_entities.items():
        values = normalize_to_list(value)
        for v in values:
            entities.append({
                "label": label,
                "text": v
            })
    
    converted.append({
        "file_name": file_name,
        "entities": entities
    })

# Speichern
with open("../../data/NER/gemma2/prompt5/gemma2_prompt5_2_as_goldstructure.json", "w", encoding="utf-8") as f:
    json.dump(converted, f, ensure_ascii=False, indent=2)

````

````python
import json
from tqdm import tqdm
from collections import defaultdict
from sklearn.metrics import precision_recall_fscore_support
import pandas as pd

# === Dateien laden ===
with open("../../data/data_annotated.json", encoding="utf-8") as f:
    gold_data = json.load(f)

with open("../../data/NER/gemma2/prompt5/gemma2_prompt5_1_as_goldstructure.json", encoding="utf-8") as f:
    gemma_data = json.load(f)

# === Zu vergleichende Labels ===
label_names = ["LOC", "DATE", "TIME", "EVENT", "TOPIC"]

# === Entitäten extrahieren ===
def extract_entities(entry):
    ent_dict = {}
    for ent in entry.get("entities", []):
        label = ent.get("label")
        text = ent.get("text", "").strip()
        if label and text:
            ent_dict.setdefault(label, []).append(text)
    return ent_dict

# === Mapping der Vorhersagen für schnellen Zugriff ===
gemma_dict = {entry["file_name"]: entry for entry in gemma_data}

# === Vergleich vorbereiten ===
y_true_all = []
y_pred_all = []
y_true_per_label = defaultdict(list)
y_pred_per_label = defaultdict(list)

# === Vergleichsloop über alle Goldstandard-Dateien ===
for gold_entry in tqdm(gold_data):
    file_name = gold_entry["file_name"]
    gold_ents = extract_entities(gold_entry)
    pred_entry = gemma_dict.get(file_name)
    pred_ents = extract_entities(pred_entry) if pred_entry else {}

    for label in label_names:
        gold_texts = set(t.strip() for t in gold_ents.get(label, []) if isinstance(t, str))
        val = pred_ents.get(label)

        if isinstance(val, list):
            pred_texts = set(t.strip() for t in val if isinstance(t, str))
        elif isinstance(val, str):
            pred_texts = {val.strip()}
        else:
            pred_texts = set()

        for text in gold_texts:
            if text in pred_texts:
                y_true_all.append(1)
                y_pred_all.append(1)
                y_true_per_label[label].append(1)
                y_pred_per_label[label].append(1)
                pred_texts.remove(text)
            else:
                y_true_all.append(1)
                y_pred_all.append(0)
                y_true_per_label[label].append(1)
                y_pred_per_label[label].append(0)

        for text in pred_texts:
            y_true_all.append(0)
            y_pred_all.append(1)
            y_true_per_label[label].append(0)
            y_pred_per_label[label].append(1)

# === Gesamtmetriken ===
precision, recall, f1, _ = precision_recall_fscore_support(
    y_true_all, y_pred_all, average="binary", zero_division=0
)

print("\n🔍 Gesamtergebnis (alle Goldstandard-Dateien)")
print(f"Precision: {precision:.3f}")
print(f"Recall:    {recall:.3f}")
print(f"F1-Score:  {f1:.3f}")

# === Pro-Label-Metriken ===
rows = []
for label in label_names:
    y_true = y_true_per_label[label]
    y_pred = y_pred_per_label[label]

    if not y_true and not y_pred:
        precision = recall = f1 = 0.0
    else:
        precision, recall, f1, _ = precision_recall_fscore_support(
            y_true, y_pred, average="binary", zero_division=0
        )

    rows.append({
        "Label": label,
        "Precision": round(precision, 3),
        "Recall": round(recall, 3),
        "F1-Score": round(f1, 3),
        "Anzahl in Gold": sum(y_true),
        "Anzahl Predicted": sum(y_pred)
    })

# Ausgabe
df_metrics = pd.DataFrame(rows).sort_values("Label")
print("\n📊 Metriken pro Kategorie:")
print(df_metrics.to_string(index=False))
````

````python
import json
from tqdm import tqdm
from collections import defaultdict
from sklearn.metrics import precision_recall_fscore_support
import pandas as pd

# === Dateien laden ===
with open("../../data/data_annotated.json", encoding="utf-8") as f:
    gold_data = json.load(f)

with open("../../data/NER/gemma2/prompt5/gemma2_prompt5_1_as_goldstructure.json", encoding="utf-8") as f:
    gemma_data = json.load(f)

# === Zu vergleichende Labels ===
label_names = ["LOC", "DATE", "TIME", "EVENT", "TOPIC"]

# === Entitäten extrahieren ===
def extract_entities(entry):
    ent_dict = {}
    for ent in entry.get("entities", []):
        label = ent.get("label")
        text = ent.get("text", "").strip()
        if label and text:
            ent_dict.setdefault(label, []).append(text)
    return ent_dict

# === Gemeinsame Dateien filtern ===
gold_files = {entry["file_name"] for entry in gold_data}
gemma_files = {entry["file_name"] for entry in gemma_data}
common_files = gold_files & gemma_files

filtered_gold = [entry for entry in gold_data if entry["file_name"] in common_files]
gemma_dict = {entry["file_name"]: entry for entry in gemma_data if entry["file_name"] in common_files}

# === Vergleich vorbereiten ===
y_true_all = []
y_pred_all = []

# pro Label getrennte Zählung
y_true_per_label = defaultdict(list)
y_pred_per_label = defaultdict(list)

# === Vergleichsloop ===
for gold_entry in tqdm(filtered_gold):
    file_name = gold_entry["file_name"]
    gold_ents = extract_entities(gold_entry)
    pred_entry = gemma_dict.get(file_name)
    pred_ents = extract_entities(pred_entry) if pred_entry else {}

    for label in label_names:
        gold_texts = set(t.strip() for t in gold_ents.get(label, []) if isinstance(t, str))
        val = pred_ents.get(label)

        if isinstance(val, list):
            pred_texts = set(t.strip() for t in val if isinstance(t, str))
        elif isinstance(val, str):
            pred_texts = {val.strip()}
        else:
            pred_texts = set()

        # True Positives und False Negatives
        for text in gold_texts:
            if text in pred_texts:
                y_true_all.append(1)
                y_pred_all.append(1)
                y_true_per_label[label].append(1)
                y_pred_per_label[label].append(1)
                pred_texts.remove(text)
            else:
                y_true_all.append(1)
                y_pred_all.append(0)
                y_true_per_label[label].append(1)
                y_pred_per_label[label].append(0)

        # Verbleibende Vorhersagen = False Positives
        for text in pred_texts:
            y_true_all.append(0)
            y_pred_all.append(1)
            y_true_per_label[label].append(0)
            y_pred_per_label[label].append(1)

# === Gesamtmetriken ===
precision, recall, f1, _ = precision_recall_fscore_support(
    y_true_all, y_pred_all, average="binary", zero_division=0
)

print("\n🔍 Gesamtergebnis (nur gemeinsame Dateien)")
print(f"Precision: {precision:.3f}")
print(f"Recall:    {recall:.3f}")
print(f"F1-Score:  {f1:.3f}")

# === Metriken pro Label ===
rows = []
for label in label_names:
    y_true = y_true_per_label[label]
    y_pred = y_pred_per_label[label]

    if not y_true and not y_pred:
        precision = recall = f1 = 0.0
    else:
        precision, recall, f1, _ = precision_recall_fscore_support(
            y_true, y_pred, average="binary", zero_division=0
        )

    rows.append({
        "Label": label,
        "Precision": round(precision, 3),
        "Recall": round(recall, 3),
        "F1-Score": round(f1, 3),
        "Anzahl in Gold": sum(y_true),
        "Anzahl Predicted": sum(y_pred)
    })

# Ausgabe als Tabelle
df_metrics = pd.DataFrame(rows).sort_values("Label")
print("\n📊 Metriken pro Kategorie:")
print(df_metrics.to_string(index=False))
````

````python
import json
from tqdm import tqdm
from collections import defaultdict
from sklearn.metrics import precision_recall_fscore_support
import pandas as pd

# === Dateien laden ===
with open("../../data/data_annotated.json", encoding="utf-8") as f:
    gold_data = json.load(f)

with open("../../data/NER/gemma2/prompt5/gemma2_prompt5_2_as_goldstructure.json", encoding="utf-8") as f:
    gemma_data = json.load(f)

# === Zu vergleichende Labels ===
label_names = ["LOC", "DATE", "TIME", "EVENT", "TOPIC"]

# === Entitäten extrahieren ===
def extract_entities(entry):
    ent_dict = {}
    for ent in entry.get("entities", []):
        label = ent.get("label")
        text = ent.get("text", "").strip()
        if label and text:
            ent_dict.setdefault(label, []).append(text)
    return ent_dict

# === Gemeinsame Dateien filtern ===
gold_files = {entry["file_name"] for entry in gold_data}
gemma_files = {entry["file_name"] for entry in gemma_data}
common_files = gold_files & gemma_files

filtered_gold = [entry for entry in gold_data if entry["file_name"] in common_files]
gemma_dict = {entry["file_name"]: entry for entry in gemma_data if entry["file_name"] in common_files}

# === Vergleich vorbereiten ===
y_true_all = []
y_pred_all = []

# pro Label getrennte Zählung
y_true_per_label = defaultdict(list)
y_pred_per_label = defaultdict(list)

# === Vergleichsloop ===
for gold_entry in tqdm(filtered_gold):
    file_name = gold_entry["file_name"]
    gold_ents = extract_entities(gold_entry)
    pred_entry = gemma_dict.get(file_name)
    pred_ents = extract_entities(pred_entry) if pred_entry else {}

    for label in label_names:
        gold_texts = set(t.strip() for t in gold_ents.get(label, []) if isinstance(t, str))
        val = pred_ents.get(label)

        if isinstance(val, list):
            pred_texts = set(t.strip() for t in val if isinstance(t, str))
        elif isinstance(val, str):
            pred_texts = {val.strip()}
        else:
            pred_texts = set()

        # True Positives und False Negatives
        for text in gold_texts:
            if text in pred_texts:
                y_true_all.append(1)
                y_pred_all.append(1)
                y_true_per_label[label].append(1)
                y_pred_per_label[label].append(1)
                pred_texts.remove(text)
            else:
                y_true_all.append(1)
                y_pred_all.append(0)
                y_true_per_label[label].append(1)
                y_pred_per_label[label].append(0)

        # Verbleibende Vorhersagen = False Positives
        for text in pred_texts:
            y_true_all.append(0)
            y_pred_all.append(1)
            y_true_per_label[label].append(0)
            y_pred_per_label[label].append(1)

# === Gesamtmetriken ===
precision, recall, f1, _ = precision_recall_fscore_support(
    y_true_all, y_pred_all, average="binary", zero_division=0
)

print("\n🔍 Gesamtergebnis (nur gemeinsame Dateien)")
print(f"Precision: {precision:.3f}")
print(f"Recall:    {recall:.3f}")
print(f"F1-Score:  {f1:.3f}")

# === Metriken pro Label ===
rows = []
for label in label_names:
    y_true = y_true_per_label[label]
    y_pred = y_pred_per_label[label]

    if not y_true and not y_pred:
        precision = recall = f1 = 0.0
    else:
        precision, recall, f1, _ = precision_recall_fscore_support(
            y_true, y_pred, average="binary", zero_division=0
        )

    rows.append({
        "Label": label,
        "Precision": round(precision, 3),
        "Recall": round(recall, 3),
        "F1-Score": round(f1, 3),
        "Anzahl in Gold": sum(y_true),
        "Anzahl Predicted": sum(y_pred)
    })

# Ausgabe als Tabelle
df_metrics = pd.DataFrame(rows).sort_values("Label")
print("\n📊 Metriken pro Kategorie:")
print(df_metrics.to_string(index=False))
````

#### Halluziniert das Modell?

````python
import json
from tqdm import tqdm
from collections import defaultdict, Counter

# === Dateien laden ===
with open("../../data/data_annotated.json", encoding="utf-8") as f:
    gold_data = json.load(f)

with open("../../data/NER/gemma2/prompt5/gemma2_prompt5_2_as_goldstructure.json", encoding="utf-8") as f:
    gemma_data = json.load(f)

# === Zu vergleichende Labels ===
label_names = ["LOC", "DATE", "TIME", "EVENT", "TOPIC"]

# === Entitäten extrahieren ===
def extract_entities(entry):
    ent_dict = {}
    for ent in entry.get("entities", []):
        label = ent.get("label")
        text = ent.get("text", "").strip()
        if label and text:
            ent_dict.setdefault(label, []).append(text)
    return ent_dict

# === Gemeinsame Dateien filtern ===
gold_files = {entry["file_name"] for entry in gold_data}
gemma_files = {entry["file_name"] for entry in gemma_data}
common_files = gold_files & gemma_files

filtered_gold = [entry for entry in gold_data if entry["file_name"] in common_files]
gemma_dict = {entry["file_name"]: entry for entry in gemma_data if entry["file_name"] in common_files}

# === Halluzinationen prüfen ===
hallucinated_entities = defaultdict(list)
total_hallucinations = 0
label_counter = Counter()

for gold_entry in tqdm(filtered_gold):
    file_name = gold_entry["file_name"]
    text = gold_entry.get("text", "").lower()
    pred_entry = gemma_dict.get(file_name)
    pred_ents = extract_entities(pred_entry) if pred_entry else {}

    for label in label_names:
        for ent in pred_ents.get(label, []):
            ent_str = ent.strip()
            if ent_str and ent_str.lower() not in text:
                hallucinated_entities[file_name].append({
                    "label": label,
                    "hallucinated_entity": ent_str
                })
                total_hallucinations += 1
                label_counter[label] += 1

# === Ergebnis ausgeben ===
print(f"\n🔍 Gesamtanzahl halluzinierter Entitäten: {total_hallucinations}")

print("\n📉 Halluzinationen nach Label:")
for label, count in label_counter.items():
    print(f"  {label}: {count}")

# Optional: Beispiele anzeigen
print("\n📄 Beispiele halluzinierter Entitäten (max. 5 Dateien):")
for file_name, ents in list(hallucinated_entities.items())[:5]:
    print(f"\nDatei: {file_name}")
    for e in ents:
        print(f"  ❌ [{e['label']}] \"{e['hallucinated_entity']}\" kommt nicht im Text vor.")
````

#### Anzahl der exakten Übereinstimmungen (True Positives) je Entität


````python
import json
from collections import defaultdict

with open("../../data/data_annotated.json", encoding="utf-8") as f:
    gold = json.load(f)

with open("../../data/NER/gemma2/prompt5/gemma2_prompt5_1_as_goldstructure.json", encoding="utf-8") as f:
    ner = json.load(f)

# Erzeuge Lookup nach file_name für schnellen Zugriff
gold_by_file = {entry["file_name"]: entry for entry in gold}
ner_by_file = {entry["file_name"]: entry for entry in ner}

# Zähler für genaue Übereinstimmungen pro Label
match_counts = defaultdict(int)
total_counts = defaultdict(int)

def get_entities(entry):
    """Erzeuge ein Dict {label: set of texts} für eine Datei"""
    entities = defaultdict(set)
    for e in entry.get("entities", []):
        label = e["label"]
        text = e["text"].strip().lower()
        entities[label].add(text)
    return entities

# Iteriere über alle Dateien im Goldstandard
for file_name, gold_entry in gold_by_file.items():
    ner_entry = ner_by_file.get(file_name)
    if ner_entry is None:
        continue  # Kein Output vorhanden für diese Datei

    gold_entities = get_entities(gold_entry)
    ner_entities = get_entities(ner_entry)

    # Betrachte alle Labels, die vorkommen
    all_labels = set(gold_entities.keys()) | set(ner_entities.keys())
    for label in all_labels:
        gold_set = gold_entities.get(label, set())
        ner_set = ner_entities.get(label, set())

        # Zähle Totalfälle (Gold)
        total_counts[label] += len(gold_set)

        # Zähle nur exakte Übereinstimmungen
        matches = gold_set & ner_set
        match_counts[label] += len(matches)

# Ergebnis als dict
results = {label: {"matches": match_counts[label], "total": total_counts[label]} for label in total_counts}
results
````
