### Test der Rechtschreibkorrektur auf einzelnen Daten - stichprobenartig

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("oliverguhr/spelling-correction-german-base")
model = AutoModelForSeq2SeqLM.from_pretrained("oliverguhr/spelling-correction-german-base")
```

```python
import pandas as pd

df_ocr = pd.read_csv("../../../data/OCR/easyocr_evaluation_01_0.csv")
df_ocr.head(1)
```

```python
df_ocr = df_ocr[["file_name","ocr_easyocr"]]
df_ocr.head(1)
```

#### TEST 1

```python
text = df_ocr[df_ocr["file_name"] == "0001.jpg"]["ocr_easyocr"].values[0]
print(text)
```

```python
from transformers import pipeline

fix_spelling = pipeline("text2text-generation",model="oliverguhr/spelling-correction-german-base", device=-1) #device=-1 damit es auf CPU läuft

corrected = fix_spelling(text,max_length=256)
print(corrected)
```
