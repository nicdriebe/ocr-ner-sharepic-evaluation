## Flair Evaluation - 01/1 - qualitative Analyse

Es wird folgendes betrachtet: 
- True Postives / Beispiele
- False Postives  / Anzahl, Beispiele, Fehler-Kategorien
- False Negatives  / Anzahl



zusätzlich: Einfluss der Textlänge auf LOC Erkennung (quantitative Betrachtung)


---

#### true positives

```python
import pandas as pd
import json
from IPython.display import display


with open("../../data/NER/flair/results_ner_german_large.json", encoding="utf-8") as f:
    pred_data = json.load(f)

with open("../../data/data_annotated.json", encoding="utf-8") as f:
    gold_data = json.load(f)

# 2. Golddaten nach file_name indexieren
gold_index = {entry["file_name"]: entry for entry in gold_data}
relevant_label = "LOC"

# 3. TP sammeln + Goldvergleich einbauen
true_positives_all = []

for pred in pred_data:
    file_name = pred["file_name"]
    text = pred["text"]
    tps = [e for e in pred.get("true_positives", []) if e["label"] == relevant_label]

    gold_entry = gold_index.get(file_name, {})
    gold_locs = [
        {
            "start": e["start"],
            "end": e["end"],
            "label": e["label"],
            "text": text[e["start"]:e["end"]]
        }
        for e in gold_entry.get("entities", [])
        if e["label"] == relevant_label
    ]

    for tp in tps:
        start, end = tp["start"], tp["end"]
        tp_entry = {
            "file_name": file_name,
            "text_fragment": text[start:end],
            "start": start,
            "end": end,
            "label": tp["label"],
            "full_text": text,
            "gold_locs": gold_locs  
        }
        true_positives_all.append(tp_entry)


df_tp = pd.DataFrame(true_positives_all)


df_tp = df_tp[["file_name", "text_fragment", "start", "end", "gold_locs"]]

pd.set_option('display.max_colwidth', None)


df_tp.head(10)
```

### Für genauere Analysen wurden im NER-Json bei "false positives" manuell weitere Felder angelegt:  

**loc_valid: true** (wenn LOC eigentlich richtig aber nicht relevant für die Vernastaltung)  
**partial_match: true** (wenn LOC nur partiell erkannt wurde, also zum Beispiel die Hausnummer fehlte)   
**merged_match: true** (wenn LOC aus mehreren LOC Entitäten des Goldstandards besteht.)   
**split_entity: true** (wenn LOC im Goldstandard aus merhen Wörtern besteht und im NER alle Wörter vorhanden sind, aber getrennt ausgegeben wurden.)  
**correct_label:  ...** (wird nur gesetzt, wenn eigentlich eine andere sinnvolle Entitäte erkannt worden ist, z.B. DATE)

Felder wurden nur gesetzt wenn Wert true oder Inhalt bei correct_label vorhanden, ansonsten nicht.

#### Anzahl von LOC in false positive

```python

import json

with open("../../data/NER/flair/results_ner_german_large.json", encoding="utf-8") as f:
    pred_data = json.load(f)


# Alle False Positives mit label=LOC extrahieren
false_positives = [
    fp for entry in pred_data
    for fp in entry.get("false_positives", [])
    if fp.get("label") == "LOC"
]

# 1. Gesamtanzahl False Positives
total_fp = len(false_positives)

# 2. loc_valid: true
valid_loc = sum(1 for fp in false_positives if fp.get("loc_valid") is True)

# 3. partial_match: true
partial = sum(1 for fp in false_positives if fp.get("partial_match") is True)

# 4. merged_gold_entities: true
merged = sum(1 for fp in false_positives if fp.get("merged_match") is True)

# 5. split_entity: true
split = sum(1 for fp in false_positives if fp.get("split_entity") is True)

# 6. correct_label gesetzt
correct_label_set = sum(1 for fp in false_positives if "correct_label" in fp and fp["correct_label"] not in [None, "", "NONE"])

# 7. keine Zusatzinfo gesetzt (nichts von oben)
no_annotation = sum(
    1 for fp in false_positives
    if not any(fp.get(key) for key in ["loc_valid", "partial_match", "merged_match", "split_entity", "correct_label"])
)

# Ausgabe
print(f"1. Gesamtzahl False Positives (LOC): {total_fp}")
print(f"2. loc_valid = true               : {valid_loc}")
print(f"3. partial_match = true           : {partial}")
print(f"4. merged_matched = true          : {merged}")
print(f"5. split_entity = true            : {split}")
print(f"6. correct_label gesetzt          : {correct_label_set}")
print(f"7. keine Zusatzinfo gesetzt       : {no_annotation}")
```

274 False Positives, davon sind:

**106 LOC Entitäten wurden teilweise erkannt** -> in den allermeistenfällen fehlte entweder die Postleitzahl oder Hausnummer

**102 LOC Entitäten weisen nicht auf Vernastltungsort hin** - sie können LOC zugeordnet werden, aber passen inhaltich nicht zum Veranstaltungsort passen. Es sind Orte, die im Slogan, Thema vorkommen oder als Teil von Organisationsnamen

**8 LOC Entitäten wurden vom Modell zusammengefasst**, waren im Goldstandard getrennt annotiert und wurden deshalb nicht erkannt, waren aber vollständig

**3 LOC Entitäten wurden vom Modell aufgesplittet**, aber waren im Goldstandard als eine zusammenhängende Entität gesetzt. Sie waren ebenfalls vollständig und richtig.

**52 Entitäten wurden falsch als LOC annotiert**, sind keine geografischen Orte

von den insgesamt 274 LOCs sind somit 52 komplett fehlerhaft erkannt worden

## Textlänge vs LOC-Erkennung

```python
import pandas as pd
import json
from IPython.display import display


with open("../../data/NER/flair/results_ner_german_large.json", encoding="utf-8") as f:
    pred_data = json.load(f)

data = []

for pred in pred_data:
    text = pred["text"]
    file_name = pred["file_name"]
    text_len = len(text)

    tps = [e for e in pred.get("true_positives", []) if e["label"] == "LOC"]
    fps = [e for e in pred.get("false_positives", []) if e["label"] == "LOC"]
    fns = [e for e in pred.get("false_negatives", []) if e["label"] == "LOC"]

    tp_count = len(tps)
    fp_count = len(fps)
    fn_count = len(fns)

    precision = tp_count / (tp_count + fp_count) if (tp_count + fp_count) > 0 else 0
    recall = tp_count / (tp_count + fn_count) if (tp_count + fn_count) > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    data.append({
        "file_name": file_name,
        "text_length": text_len,
        "tp": tp_count,
        "fp": fp_count,
        "fn": fn_count,
        "precision": precision,
        "recall": recall,
        "f1": f1
    })

import pandas as pd
df = pd.DataFrame(data)

df.head()
```

```python
df["length_group"] = pd.cut(df["text_length"], bins=[0, 100, 300, 500, 800, 1200, 1600])
df.groupby("length_group", observed=True)["f1"].mean()

```
